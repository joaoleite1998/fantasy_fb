```python
import requests
import bs4
from tqdm import tqdm

def scrape_fantasy_pts(url):
    res = requests.get(url, headers={"User-Agent":"Mozilla/5.0"})
    soup = bs4.BeautifulSoup(res.text, 'html.parser')
    table_block = soup.find('table', class_='statistics scrollable')

    data_rows = table_block.find_all('tr', class_=['row0 right', 'row1 right'])
    print("------------------------------------------------------------")
    for row in data_rows:
        player_name = row.a.string

        fantasy_pts_block = row.find('td', class_='hilite')
        fantasy_pts = fantasy_pts_block.string
        print(player_name, fantasy_pts)
        

urls = ['https://www.footballdb.com/fantasy-football/index.html?pos=QB%2CRB%2CWR%2CTE&yr=2019&wk=all&rules=1',
        'https://www.footballdb.com/fantasy-football/index.html?pos=K&yr=2019&wk=all&rules=1']

for url in urls:
    scrape_fantasy_pts(url)
    
```
